#### Task:
- Read the data again
- Apply for loop
- Iterate columns
- Get the unique labels for each column
- Get the count of unique labels
- Create a dictionary
- Apply in to the map

```python
visa_df=pd.read_csv(r"C:\Users\MOHD RAHEEM\Documents\Naresh IT\Data-Files\Visadataset.csv")
visa_df['case_status'].unique()
```

```python
cols=visa_df.select_dtypes(include='object').columns
for i in cols:
    labels=list(visa_df[i].unique())
    values=[i for i in range(len(labels))]
    d=dict(zip(labels,values))
    visa_df[i]=visa_df[i].map(d)
visa_df
```

#### One hot encoder:
- One hot means if one is ON another is OFF
- ON represents with 1, & OFF represents with 0
- For example case_status has two unique labels
    - Denied 
    - Certified
- If we apply one hot encoder on case_status column it will create two extra columns
    - case_status_Denied
    - case_status_Certified

|case_status|case_status_denied|case_status_certified
|------|--|--|
|Denied|1|0|
|Certified|0|1|

#### Advantages:
- In machine learning we have one important property is **Independent of variable**
- ML models expects variables are **independent each other**
- One hot coder creating new columns, the new columns are **independent each other**
- Independent variable means both variables have **90 degrees phase shift**
- 90 degrees phase shift also called as **perpendicular each other**
- Perpendicular means **orthogonal each other**

**Disadvantage**
- One hot encoder creates new columns equal to number of unique labels
- If number of unique labels increase, new columns also get increase
- The more columns leads to more data
- If the data is more complexity increase
- Complexity increses computation power required
- Also time also increase 
- The entire process is named as **curse of dimensionality**
