#### Data Transformations:
- Generally data has 3 types of distributions
    - Normal distribution
    - Positive Skewed 
    - Negative Skewed 
- All the maths concepts developed by assumption as Data follows Normal distribution
- Data skew happens because of many reasons, in that one reason is outliers
- Once we perform the outlier analysis then we can apply transformation methods, if the data does not follows normality
- We have some transformation methods mentioned below
    - Log Transformation
    - Exponential Transformation
    - Reciprocal Transformation
    - Square root Transformation
    - Power Transformation

#### Step1: Read the packages:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Step-2: **Read the data:**

```python
exp_data=np.random.exponential(size=10000)
exp_data

# random.exponential() means wantedly we're taking random data,
# Because we know that the random data does not has the normal distribution plot.
```

#### Step3:
- Plot the data

```python
plt.hist(exp_data,bins=20,label='Exponential') # bins == intervals
plt.show()
```

```python
norm_data=np.random.normal(size=10000)
plt.hist(norm_data,bins=20,label='Normal')
plt.show()

# Now here we're considering noraml data by default
# Usind random.normal()
```

#### Goal
- **Convert exp_data(exponentaial) to Normal distribution**
- bins means interval

#### Step-4:
- Log Transformation
- Log Transformation means performin logarithm operations on original data
- It is an approach to convert into Normal distribution
- Log means natural logarithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data,bins=20,label='Logarithm')  # bins == intervals
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(exp_data,bins=30,label='Exponential')  # random data(non_normal)
plt.legend()
plt.subplot(1,2,2)
plt.hist(log_data,bins=30,label='Logarithm') 
plt.legend()
plt.show()
```

#### Reciprocal Transformations:

- Reciprocal fails when values are very low
- For examplr x=0 then 1/0 == fails

```python
x=2
1/x
```

```python
reciprocal_data=np.reciprocal(exp_data)
reciprocal_data
```

```python
reciprocal_data=np.reciprocal(exp_data)
plt.hist(reciprocal_data,bins=20,label='Reciprocal')  # bins == intervals
plt.legend()
plt.show()
```

#### Square root transformation:

```python
sqrt_data=np.sqrt(exp_data)
sqrt_data
```

```python
sqrt_data=np.sqrt(exp_data)
plt.hist(sqrt_data,bins=20,label='Square_root')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3))

plt.subplot(1,3,1)
plt.hist(exp_data,bins=30,label='Exponential')  # original(non-normal)
plt.legend()

plt.subplot(1,3,2)
plt.hist(log_data,bins=30,label='Logarithm')    # applying np.log(original(non-normal))
plt.legend()

plt.subplot(1,3,3)
plt.hist(sqrt_data,bins=30,label='Square_root')  # applying np.sqrt(original(non-normal))
plt.legend()
plt.show()
```

#### Power Transformer:
- box-cox
- yeo-jhonson
- It is under sklearn preprocessing

```python
#Step1: Read the package:
from sklearn.preprocessing import PowerTransformer

# Step-2: Save the package
power=PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)

# Step-3: Apply fit transform 
data_trans=power.fit_transform(exp_data1)

plt.hist(data_trans)
```
